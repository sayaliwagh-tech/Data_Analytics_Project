```python
import pandas as pd

df = pd.read_csv('cleaned_superstore.csv')
```

```python
# Total Sales and Profit
'Total Sales :', df['sales'].sum()
'Total Profit :', df['profit'].sum()
```

```python
# Sales by region
df.groupby('region')['sales'].sum().sort_values(ascending=False)
```

```python
# Profit by category
df.groupby('category')['profit'].sum().sort_values(ascending=False)
```

```python
# Top 10 states by sales
df.groupby('state')['sales'].sum().sort_values(ascending=False).head(10)
```

```python
# Average profit by discount
df.groupby('discount')['profit'].mean().sort_values()
```

```python
# Sub-categories by profit
df.groupby('sub_category')['profit'].sum().sort_values(ascending=False)
```

```python
# Sales by customer segment
df.groupby('segment')['sales'].sum().sort_values(ascending=False)
```

```python
# Profit by customer segment
df.groupby('segment')['sales'].sum().sort_values(ascending=False)
```

```python
# Average profit by discount
df.groupby('discount')['profit'].mean().sort_values()
```
